```python
# Importamos pyplot de matplotlib
import matplotlib.pyplot as plt
```

```python
# Importamos numpy
import numpy as np
```

```python
# Hacemos un arreglo desde 0 hasta 3 * PI con incremento de 0.1
x = np.arange(0, 3 * np.pi, 0.1)
x
```

```python
# Ahora para la coordenada en y hacemos el seno de cada elememnto
y = np.sin(x)
y
```

```python
# Le ponemos un titulo al grafico
plt.title("Grafica de la funcion seno")
# Graficamos los puntos
plt.plot(x,y)
# Mostramos la grafica
plt.show()
```

```python
# Ahora haremos lo mismo pero con colores en la grafica y guardandola en pdf
#Creamos un array en numpy
x = np.arange(0,100)
x
```

```python
# Ahora 'y' la elevaremos al cuadrado 
y = x ^ 10
y
```

```python
# Titulo de la grafica
plt.title("Costo de productos")
# Etiqueta de eje x
plt.xlabel("Costo")
# Etiqueta de eje y
plt.ylabel("Productos")
# Generamos la grafica y le damos color azul
plt.plot(x,y,'b')
# Ahora la guardamos en jpg
plt.savefig("costo_productos.jpg", format="jpg")
# Mostramos la grafica
plt.show()
```
