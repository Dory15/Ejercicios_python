```python
# Por ejemplo para aplicar una funcion a un data frame
# Que lo que va a hacer es agregar una columna a el data frame con la inicial de cada camara
df_camaras_1["letra_camara"] = df_camaras_1.Model.apply(lambda nombre: nombre[0])
```

```python
# Ahora vamos a mostralo 
df_camaras_1.head(20)
```

```python
# Ahora vamos a borrar la columna que hicimos
# El 0 es para borrar filas y el 1 para borrar columnas
# El inplace es para no tener que asignarlo a otra variable
df_camaras_1.drop("letra_camara", axis=1, inplace=True)
df_camaras_1
```

```python
# Vamos a mostrar otro archivo csv
!type csv_camaras_2.csv
```

```python
df_camaras_2 = pd.read_csv("csv_camaras_2.csv")
df_camaras_2
```

```python
# Tambien podemos unir dataframes como un JOIN en MySQL
# Left_on es para definir la columna con la que queremos unirlo del data frame que esta afuera del parentesis
# Mientras que right_on es para definir la columna con la que queremos unirlo del data frame que esta dentro del parentesis
df_merged = df_camaras_1.merge(df_camaras_2, left_on="Model", right_on="Model")
df_merged
```

```python
# Por ultimo podemos exportar el data frame a una archivo excel
df_merged.to_excel("camaras.xls", sheet_name="datos")
```

```python
# Tambien podemos exportarlo a un archivo csv
df_merged.to_csv("camaras.csv")
```
